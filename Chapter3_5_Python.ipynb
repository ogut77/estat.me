```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data file
url = 'https://raw.githubusercontent.com/ogut77/DataScience/refs/heads/main/PurchaseByCredit40.csv'
card = pd.read_csv(url)


```

```python
# Display the data
print(card)
```

```python

# 1 dim frequency table
card['Gender'].value_counts()
```

```python
 # 2 dim frequency table

pd.crosstab(card['Gender'], card['Purchase'])
```

```python
# prompt: # 3 dim frequency table
# table(Gender,Age,Purchase)

pd.crosstab([card['Gender'], card['Age']], card['Purchase'])
```

```python


import pandas as pd

# Load Iris dataset from UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width', 'Species']
iris= pd.read_csv(url, header=None, names=columns)
#Display the data
iris

```

```python

# mean of 'Sepal.Length'
iris['Sepal.Length'].mean()

```

```python
# median of 'Sepal.Length'
iris['Sepal.Length'].median()
```

```python
# variance of 'Sepal.Length'
iris['Sepal.Length'].var()
```

```python
# standard deviation of 'Sepal.Length'
iris['Sepal.Length'].std()
```

```python

# range of 'Sepal.Length'
print(iris['Sepal.Length'].min(), iris['Sepal.Length'].max())
```

```python
#Summary of basic statistics
iris.describe()
```

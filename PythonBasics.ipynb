# **Arithmetic Operations**

```python
# Python as a simple calculator for arithmetic operations
# Operators: + (addition), - (subtraction), * (multiplication), / (division), ** (exponentiation)
# Order of precedence: ** (highest), then * and /, then + and -
# Spaces are ignored, and # indicates a comment

# Examples of arithmetic operations
print(3 + 8)      # Addition, Output: 11
print(3 - 5)      # Subtraction, Output: -2
print(8 + 3 * 5)  # Multiplication first, Output: 23
print(1 + 6 / 3)  # Division first, Output: 3
print(3**2 + 2)   # Exponentiation first, Output: 11

```

```python
# Python's modular operator % calculates the remainder
# Example: remainder of 24 divided by 7
print(24 % 7)  # Output: 3
```

# **Stroge of Number As Variables**



```python
# Storing numbers in variables and performing calculations
a = 3  # Store 3 in variable a
b = 5  # Store 5 in variable b
c = a + b  # Add a and b, store result in c
print(c)  # Output: 8
```

# **Vectors**

```python

import numpy as np

# Define a numeric vector (array in Python)
x = np.array([1, 3, 5])  # Equivalent to R's c(1, 3, 5)
print(x)  # Output: [1 3 5]

# Define a character vector
s = np.array(['a', 'b', 'c'])  # Equivalent to R's c("a", "b", "c")
print(s)  # Output: ['a' 'b' 'c']

# Create a sequence of numbers (like R's 5:10)
y = np.arange(5, 11)  # From 5 to 10 (11 is exclusive)
print(y)  # Output: [ 5  6  7  8  9 10]

# Combine two vectors (concatenate)
z = np.concatenate((x, y))  # Equivalent to R's c(x, y)
print(z)  # Output: [ 1  3  5  5  6  7  8  9 10]

# Access specific elements (0-based indexing in Python vs 1-based in R)
print(y[1])  # Second element, Output: 6
print(y[[1, 3]])  # Second and fourth elements, Output: [6 8]

# Vector operations
print(x * 3)  # Multiply each element by 3, Output: [ 3  9 15]
w = x - 3  # Subtract 3 from each element, assign to w
print(w)  # Output: [-2  0  2]
print(x**2)  # Square each element, Output: [ 1  9 25]
```

# **Matrixs and Arrays**

```python
import numpy as np

# Define a 2x3 matrix (equivalent to R's matrix(1:6, nrow=2, ncol=3))
m = np.array([[1, 3, 5], [2, 4, 6]])  # 2 rows, 3 columns
# Alternatively: m = np.arange(1, 7).reshape(2, 3)
print(m)
# Output:
# [[1 3 5]
#  [2 4 6]]

# Access matrix elements (0-based indexing in Python vs 1-based in R)
print(m[0, 1])  # 1st row, 2nd column element, Output: 3
print(m[0, :])  # All elements in 1st row, Output: [1 3 5]
print(m[:, 0])  # All elements in 1st column, Output: [1 2]

# Define a 3D array (equivalent to R's array(1:24, c(3,4,2)))
a = np.arange(1, 25).reshape(3, 4, 2)  # 3 rows, 4 columns, 2 layers
print('Printing 3D array')
print(a)
```

# **Data Frame and Basic Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd

# Create the DataFrame
test = pd.DataFrame({
    'name': ['kim', 'lee', 'park'],
    'age': [28, 33, 41],
    'income': [200, 300, 400]
})

# Display the DataFrame
print(test)
#    name  age  income
# 0   kim   28     200
# 1   lee   33     300
# 2  park   41     400

# List column names (equivalent to R's names())
print(test.columns.tolist())
# Output: ['name', 'age', 'income']

# In Python, no equivalent to R's attach() is needed; access columns directly with test['column_name']

# Calculate the mean of numeric columns (equivalent to R's mean(test))
print(test[['age', 'income']].mean())
# Output:
# age       34.0
# income   300.0
# Note: 'name' is excluded as it's non-numeric, avoiding NA-like warnings

# Create a scatter plot of age vs income (equivalent to R's plot(age, income))
plt.scatter(test['age'], test['income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs Income')
plt.show()  # Displays the plot

```
